Question 1 : Explain the difference between Absolute, Relative, and Mixed Cell


Ans: 1.**Relative Cell Reference**

* Changes row and column when copied.

* Excel adjusts the formula based on the new position.

Example :  =A2+B2

* You want the formula to adapt for each row or column.

* Most calculations (sales totals, profit per row).

2. **Absolute Cell Reference ($A$1)**

* Row and column are locked.

* Does not change when copied.

If you copy C2 down, the formula remains:

=B3*$B$1

Referring to a fixed value (tax rate, discount %, conversion factor).

3. **Mixed Cell Reference**

Locks either the row or the column (not both).

Question 2 : What is a Macro in Excel? How does it help in automation?


Ans: A Macro in Excel is a recorded or written set of instructions (VBA code) that tells Excel to perform tasks automatically.

**In simple words**:
A macro repeats your actions for you with one click or shortcut.

Macros are written in VBA (Visual Basic for Applications).


* Excel records your actions (clicks, formulas, formatting, steps) and converts them into VBA code.
Later, running the macro replays the same steps instantly.

**Example**:

Every day you:

Remove blank rows

Apply headers in bold

Format numbers

Insert total row

Time taken: 10 minutes daily

**With a Macro**:

Click one button

All steps run automatically

Time taken: 2 seconds

Question 3 : What are Text Functions in Excel? Mention any five with examples.


Ans: Text functions in Excel are used to manipulate, clean, extract, combine, or format text data.
They are very useful when working with names, codes, IDs, emails, addresses, etc.

**Common Text Functions (with Examples)**

1. LEFT ()

=LEFT("ExcelFunctions",5)

Output : Excel

2.RIGHT()

Extracts characters from the right side of a text.

=RIGHT("Invoice_2024",4)
Output: 2024

3.MID()

=MID("EMP-1025-IN",5,4)
Output: 1025

4.LEN()

Returns the number of characters in a text.

=LEN("Excel")
Output: 5

Use case: Data validation, checking ID length.

5.CONCAT() / CONCATENATE()

Combines text from multiple cells.

=CONCAT("Data"," ","Analyst")
Output: Data Analyst

Use case: Full name creation, address formatting.

Question 5 : Define the purpose of VLOOKUP and HLOOKUP. How are they different
from XLOOKUP? Which among XLOOKUP and INDEX-MATCH is best while usage?


Ans: **VLOOKUP (Vertical Lookup)**

Searches for a value in the first column of a table and returns a value from a specified column number.

**Example:**

=VLOOKUP(102, A2:C10, 3, FALSE)


**HLOOKUP (Horizontal Lookup)**

Searches for a value in the first row and returns data from a specified row number.

=HLOOKUP("Sales", A1:F5, 4, FALSE)

**XLOOKUP (Modern Replacement)**

Searches in any direction and returns values from any range.

=XLOOKUP(102, A2:A10, C2:C10, "Not Found")


Which is best: **XLOOKUP or INDEX-MATCH**

**INDEX-MATCH Example:**

=INDEX(C2:C10, MATCH(102, A2:A10, 0))

Best in most cases:

Best for enterprise & legacy systems:

VLOOKUP and HLOOKUP are used to retrieve data vertically and horizontally based on a lookup value. However, they have limitations such as fixed lookup direction and dependency on column or row numbers. XLOOKUP overcomes these issues by offering flexible lookup direction, built-in error handling, and safer formulas. While XLOOKUP is preferred for modern Excel usage, INDEX-MATCH is still widely used for compatibility and complex lookup scenarios.
